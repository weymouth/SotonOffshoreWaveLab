## Tutorial 3: Frequency analysis

This tutorial will take a look at dynamic motions and loads using frequency analysis. 

Start by importing the same libraries.

```python
from analysis import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_rows = 20
```

Next, lets read in the wave data from tutorial 1.

```python
filename = '../WaveProbe/run2.csv'
(read_waves(filename)
 .head())
```

The `analysis.py` file has a function to take the [Fast Fourier Transform (FFT)](https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.fft.html) of a data frame. Let's apply it to a run.

```python
(read_waves(filename)
 .pipe())
```

Recall that an FFT determines the amplitude of sinusoids at every frequency which add together to reproduce the signal. The frequency bins are determined by the sampling rate and duration of the signal. Let's plot this data.

```python
waves21.query('freq<5 & freq>0').plot(x='freq',y='z',logy=True);
```

This data looks pretty noisy on a log scale since very small amplitudes can be seen clearly. However, there is a clear peak at $0.5~Hz$ with amplitude around $0.1~m$. There are also smaller peaks at $1,1.5,2$ and $3~Hz$; these are the high-order wave harmonics we discussed in lecture.

Next, lets repeat this analysis for the wave forcing on the fixed platform. 

```python
all_forces = pd.read_csv('all_forces.csv',index_col=0).groupby('run')
force21 = all_forces.get_group(21)[['time','force']].pipe(take_FFT)
ax = waves21.query('freq<5 & freq>0').plot(x='freq',y='z',logy=True);
force21.query('freq<5 & freq>0').plot(x='freq',y='force',logy=True,ax=ax);
```

The two lines have different units, so comparing their relative magnitudes is meaningless. However, the plot clearly shows that __the peak force on the piling occurs at the peak wave frequency and its harmonics__.

How about the motions? Let's take a look.

```python
all_motions = pd.read_csv('all_motion.csv',index_col=0).groupby('run')
motion21 = all_motions.get_group(21)[['time','q_m3','q_r3']].pipe(take_FFT)
ax = waves21.query('freq<5 & freq>0').plot(x='freq',y='z',logy=True)
motion21.query('freq<5 & freq>0').plot(x='freq',y='q_m3',logy=True,ax=ax)
motion21.query('freq<5 & freq>0').plot(x='freq',y='q_r3',logy=True,ax=ax);
```

Just as with the forces, the platforms are both heaving at the wave frequency and it's harmonics. 

For a linear analysis, we only care about the fundamental frequency and amplitude of each signal, which is just the FFT max value.

```python
def peak_freq(df,var):
    return df.pipe(take_FFT).pipe(find_max_row,var).freq

def peak_amp(df,var):
    return df.pipe(take_FFT).pipe(find_max_row,var)[var]

peaks = pd.DataFrame([{'run':run,
                       'freq':peak_freq(group,'z'),
                       'a':peak_amp(group,'z'),
                       'force':peak_amp(all_forces.get_group(run),'force'),
                       'q_m3':peak_amp(all_motions.get_group(run),'q_m3'),
                       'q_r3':peak_amp(all_motions.get_group(run),'q_r3')}
                      for run,group in all_waves]).set_index('run')
peaks
```

```python
plt.scatter(peaks.freq,peaks.force/peaks.a);
```
